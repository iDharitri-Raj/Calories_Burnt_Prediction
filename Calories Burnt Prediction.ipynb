#### Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

#### Data Collection and Preprocessing

```python
# Loading the data from csv file to a Pandas DataFrame
calories_data = pd.read_csv("../Datasets/calories.csv")
```

```python
# first 5 rows of the data
calories_data.head()
```

```python
# Loading the data 
exercise_data = pd.read_csv("../Datasets/exercise.csv")
```

```python
exercise_data.head()
```

##### Combining the two dataframes


```python
calories_dataset = pd.concat([exercise_data, calories_data['Calories']], axis=1)
```

```python
# first 5 rows
calories_dataset.head()
```

```python
# checking the number of rows and columns
calories_dataset.shape
```

```python
# getting some info about the data
calories_dataset.info()
```

```python
# statistical measures of the data
calories_dataset.describe()
```

```python
# converting categorical data into numerical values
calories_dataset.replace({'Gender':{'male':0, 'female':1}}, inplace=True)
```

```python
calories_dataset.head()
```

```python
# splitting features and target
X = calories_dataset.drop(columns=['User_ID', 'Calories'], axis=1)
Y = calories_dataset['Calories']
```

```python
print(X, Y)
```

#### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training
- XGBoost Regressor

```python
regressor = XGBRegressor()
```

```python
# training the XGBRegressor model with training data
regressor.fit(X_train, Y_train)
```

#### Model Evaluation

```python
# R Squared Value
training_data_prediction = regressor.predict(X_train)
r2_score_1 = metrics.r2_score(training_data_prediction, Y_train)

print("R Squared Value of training data is :", r2_score_1)

# test data accuracy
test_data_prediction = regressor.predict(X_test)
r2_score_2 = metrics.r2_score(test_data_prediction, Y_test)

print("R Squared Value of test data is :", r2_score_2)
```

```python
# Mean Absolute Error 
mae1 = metrics.mean_absolute_error(training_data_prediction, Y_train)
print("Mean Absolute Error of training data is :", mae1)

mae2 = metrics.mean_absolute_error(test_data_prediction, Y_test)
print("Mean Absolute Error of test data is :", mae2)
```
